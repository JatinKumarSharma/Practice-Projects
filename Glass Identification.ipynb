```python
x=df.drop(['Type of glass'],axis=1)
y=df['Type of glass']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=55)
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_s=s.fit_transform(x_train)
xt_s=s.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
tog=GradientBoostingRegressor(random_state=55)
tog.fit(x_s,y_train)
```

```python
tg=tog.predict(xt_s)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,tg))
print(mean_squared_error(y_test,tg))
print(root_mean_squared_error(y_test,tg))
print(r2_score(y_test,tg))
```

```python
imp=tog.feature_importances_
imfea=np.argsort(imp)[::-1]
print(imfea)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(x_s,i) for i in range(x_s.shape[1])]
print(vif)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
pg={'n_estimators':[60,80,100],'learning_rate':[0.1,0.15,0.2],'max_depth':[3,4,5],'min_samples_split':[2,3,4],'min_samples_leaf':[2,3],'subsample':[0.8,0.9,1.0],'max_features':['auto','sqrt','log2']}
```

```python
gs=GridSearchCV(estimator=tog,param_grid=pg,cv=5,n_jobs=-1,verbose=2,scoring='r2')
```

```python
gs.fit(x_s,y_train)
```

```python
print(gs.best_params_)
print(gs.best_score_)
```

```python
t=gs.best_estimator_.predict(xt_s)
```

```python
print(mean_absolute_error(y_test,t))
print(mean_squared_error(y_test,t))
print(root_mean_squared_error(y_test,t))
print(r2_score(y_test,t))
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
k=KFold(n_splits=4,shuffle=True,random_state=55)
```

```python
print(cross_val_score(tog,xt_s,y_test,cv=k,scoring='r2'))
```
