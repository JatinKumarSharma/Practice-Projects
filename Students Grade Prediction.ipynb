```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings ('ignore')
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset4/refs/heads/main/Grades.csv")
df
```

```python
df.drop(["Seat No."],axis=1,inplace=True)
df
```

```python
df.dtypes
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df=df.astype(float)
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['CGPA'].unique()
```

```python
x=df.drop(['CGPA'],axis=1)
y=df['CGPA']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=555)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
sgp=GradientBoostingRegressor(random_state=555)
sgp.fit(x_train,y_train)
```

```python
sg=sgp.predict(x_test)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,sg))
print(mean_squared_error(y_test,sg))
print(root_mean_squared_error(y_test,sg))
print(r2_score(y_test,sg))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
sgp=RandomForestRegressor(random_state=555)
sgp.fit(x_train,y_train)
```

```python
s=sgp.predict(x_test)
```

```python
print(mean_absolute_error(y_test,s))
print(mean_squared_error(y_test,s))
print(root_mean_squared_error(y_test,s))
print(r2_score(y_test,s))
```
