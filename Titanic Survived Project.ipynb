```python
ds=df.drop(['Embarked_Q','Fare','Embarked_S'],axis=1)
ds
```

```python
x=ds.drop(['Survived'],axis=1)
y=ds['Survived']
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=555)
```

```python
s=StandardScaler()
x_s=s.fit_transform(x_train)
xt_s=s.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
survive=GradientBoostingRegressor(random_state=555)
survive.fit(x_s,y_train)
```

```python
sd=survive.predict(xt_s)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,sd))
print(mean_squared_error(y_test,sd))
print(root_mean_squared_error(y_test,sd))
print(r2_score(y_test,sd))
```

```python
imp=survive.feature_importances_
imfea=np.argsort(imp)[::-1]
print(imfea)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(x_s,i) for i in range(x_s.shape[1])]
print(vif)
```
