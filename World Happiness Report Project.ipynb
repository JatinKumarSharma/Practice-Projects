```python
x=df_num.drop(['Happiness Score'],axis=1)
y=df_num['Happiness Score']
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
s=StandardScaler()
x_s=s.fit_transform(x_train)
xt_s=s.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
happy=GradientBoostingRegressor(random_state=5)
happy.fit(x_s,y_train)
```

```python
h=happy.predict(xt_s)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,h))
print(mean_squared_error(y_test,h))
print(root_mean_squared_error(y_test,h))
print(r2_score(y_test,h))
```

```python
imp=happy.feature_importances_
imfea=np.argsort(imp)[::-1]
print(imfea)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
pg={'n_estimators':[50,100,500,1000],'learning_rate':[0.05,0.1,0.2],'max_depth':[2,5,6,9],'min_samples_split':[2,5,10],'min_samples_leaf':[1,2,5],'subsample':[0.8,0.9,1.0],'max_features':['auto','sqrt','log2']}
```

```python
gs=GridSearchCV(estimator=happy,param_grid=pg,cv=5,n_jobs=-1,verbose=2,scoring='r2')
```

```python
gs.fit(x_s,y_train)
```

```python
print(gs.best_params_)
print(gs.best_score_)
```

```python
hs=gs.best_estimator_.predict(xt_s)
```

```python
print(mean_absolute_error(y_test,hs))
print(mean_squared_error(y_test,hs))
print(root_mean_squared_error(y_test,hs))
print(r2_score(y_test,hs))
```
